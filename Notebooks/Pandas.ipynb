```python
#importando as bibliotecas 
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
#criando nosso DataFrame
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
#Visualizando as 5 primeiras linhas 
df.head()
```

```python
#Quantidade de linhas e colunas 
df.shape
```

```python
#Tipos de dados 
df.dtypes
```

```python
#Receita total
df["Valor Venda"].sum()
```

```python
#Custo total
df["custo"] = df["Custo Unitário"].mul(df["Quantidade"]) #Criando a coluna de custo
```

```python
df.head()
```

```python
#Somando o custo total e arredondando em 2 casas decimais
round(df["custo"].sum(), 2)
```

```python
#Criar a coluna lucro
df["lucro"] = df["Valor Venda"] - df["custo"]
```

```python
df.head(1)
```

```python
#Total Lucro
df["lucro"].sum()
```

```python
#Criando uma coluna com total de dias para enviar o produto
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head(1)
```

```python
#AGORA QUEREMOS SABER A MÉDIA DO TEMPO DE ENVIO PARA CADA MARCA, E PARA ISSO PRECISAMOS TRANSFORMAR A COLUNA TEMPO_ENVIO EM NUMERICA

#extraindo apenas os dias 
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head(1)
```

```python
#Verificando o tipo da coluna
df["Tempo_envio"].dtype
```

```python
#Média do tempo de envio por marca
df.groupby("Marca")["Tempo_envio"].mean()
```

```python
#Verificando se temos dados faltantes
df.isnull().sum()
```

```python
#E SE A GENTE QUISER SABER O LUCRO POR ANO E MARCA?

#Agrupar por ano e marca
df.groupby([df["Data Venda"].dt.year,"Marca"])["lucro"].sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
df.groupby([df["Data Venda"].dt.year,"Marca"])["lucro"].sum()
```

```python
#Resetando o index
lucro_ano = df.groupby([df["Data Venda"].dt.year,"Marca"])["lucro"].sum().reset_index()
lucro_ano
```

```python
#Total de produtos vendidos 
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#Gráfico total de produtos vendidos 
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total Produtos Vendidos")
df.xlabel("Total")
df.ylabel("Produto");
```

```python
df.groupby(df["Data Venda"].dt.year)["lucro"].sum().plot.bar(title="Lucro por ano")
plt.xlabel("Ano")
plt.ylabel("Receita")
```

```python
#selcionando as vendas de 2019
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)["lucro"].sum().plot(title="LucroxMes")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

```python
df_2009.groupby("Marca")["lucro"].sum().plot.bar(title="LucroxMarca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```

```python
df_2009.groupby("Classe")["lucro"].sum().plot.bar(title="Lucro x Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```
